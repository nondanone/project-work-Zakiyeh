### Abstract


Hydrodynamically fully developed flow of power-law fluids under combined action of electroosmotic and pressure gradient forces in rectangular microreactors is analyzed considering heterogeneous catalytic reactions. The Poisson-Boltzmann, Cauchy momentum, and concentration equations are considered in two dimensions and after being dimensionless are numerically solved applying a finite difference algorithm. Variation of axial concentration gradient, and axial and horizontal mass diffusions are taken intoaccount as well. To accomplish a more general analysis, the velocity distribution is obtained by solving continuity and Cauchy momentum equations and is not considered as an average axial velocity for solving the concentration equation. Profiles of mean concentration of reactants are analyzed as functions of operating variables. Results reveal that mean concentration increases with increasing Péclet number and dimensionless Debye-Hückel parameter. It is also a decreasing function of Damkohler number. Presence of pressure gradient will affect the decreasing trend of mean concentration through the channel. Shear-thinning fluids, due to their nature, are most strongly influenced by pressure gradient and Newtonian and shear-thickening fluids are less affected, respectively. Favorable pressure gradient will decelerate mean concentration rate of decrease and adverse pressure gradient acts the opposite. Moreover, a larger aspect ratio leads to slower descending rate of mean concentration in purely electroosmotic and pressure-assisted flows while in a pressure-opposed flow, the behavior might change depending on the strength of pressure gradient and the value of channel aspect ratio.

### Conclusion

<div style="text-align: justify"> By means of a central finite difference numerical method, mass transfer of power-law fluids under combined electroosmotically and pressure driven flow in rectangular microreactors was studied in this paper. The present work is not limited with the assumption that axial concentration gradient in the microreactor is constant.Both mass axial and horizontal diffusions were taken into account.The velocity profile obtained by solving continuity and Cauchy momentum equations numerically is considered in the present study instead of an average axial velocity. A parametric study was performed to explore the effects of flow parameters on mean concentration. The results show that mean concentration is an increasing function of Péclet number and dimensionless Debye-Hückel parameter. Also, it decreases with increasing Damkohler number. In all cases, favorable pressure gradient slows down the descending trend of mean concentration while adverse pressure gradient does the opposite. In addition, when Gamma is positive, mean concentration decreases most gradually for shear-thinning fluids and after that are Newtonian and shear-thickening fluids, respectively. The reverse fashion is seen at negative values of Gamma. Increasing channel aspect ratio leads to slower decreasing rate of mean concentration through the microreactor in the case of favorable pressure gradient and purely electroosmotic flow. In the case of adverse pressure gradient, the behavior is not predictable from the beginning.<div>
